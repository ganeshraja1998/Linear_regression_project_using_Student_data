# Importing Libraries

```python
import numpy as np
import pandas as pd
```

# Load The Dataset

```python
data=pd.read_csv('Students.csv')
data
```

# Summarize the DataSet

```python
data.shape
```

```python
data.head()
```

# Segregate dataset into X (Input/Independent Variableas) & Y (Output/Dependent vaariables)

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
X
```

```python
Y
```

# Splitting the dataset into Train and Test

```python
#skleaarn - scikitleaarn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
x_train
```

```python
x_test
```

# Training th data (Choosing the model(Algorithm))


```python
from sklearn.linear_model import LinearRegression
stud=LinearRegression()
stud.fit(x_train,y_train)
```

# Predict for all the Data set

```python
y_predict=stud.predict(x_test)
```

```python
y_test
```

```python
y_predict
```

# Plotting

```python
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict,'ro:')
plt.title('Hours of Learning Vs Marks')
plt.xlabel('Hours of Learning')
plt.ylabel('Marks')
plt.show()
```

```python
#Perfoming Analysis
from sklearn.metrics import mean_squared_error,r2_score
rscore = r2_score(y_test,y_predict)
r=rscore*100
print(r.round(),'%')
```
